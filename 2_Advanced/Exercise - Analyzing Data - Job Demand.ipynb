```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df_ro = df[df['job_country']=='Romania'].copy()
```

```python
df_ro
```

```python
df_ro['job_posted_month'] = df_ro['job_posted_date'].dt.strftime('%B')
df_ro
```

```python
df_ro_pivot = df_ro.pivot_table(
    index = 'job_posted_month',
    columns = 'job_title_short',
    aggfunc='size')
df_ro_pivot
```

```python
df_ro_pivot.reset_index(inplace=True)
df_ro_pivot
```

```python
df_ro_pivot['month_no'] = pd.to_datetime(df_ro_pivot['job_posted_month'],format = '%B').dt.month
df_ro_pivot['month_no']
```

```python
df_ro_pivot
```

```python
df_ro_pivot.sort_values('month_no',inplace=True)
```

```python
df_ro_pivot.set_index('job_posted_month',inplace=True)
df_ro_pivot
```

```python
df_ro_pivot.drop(columns='month_no',inplace=True)
```

```python
df_ro_pivot
```

```python
df_ro_pivot.plot(kind = 'line')
```

```python
top_3 = df_ro['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
```

```python
df_ro_pivot[top_3].plot(kind = 'line')
plt.title('Monthly Job Posting for Top Data Jobs in Romania')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```
