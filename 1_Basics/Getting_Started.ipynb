```python
!pip install datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
job_counts = df.job_title_short.value_counts().sort_index()
job_counts
```

```python
plt.bar(job_counts.index, job_counts)
plt.title('Posting by Job Title')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation=45,ha = 'right') #ha = horizontal alignment
plt.show()
```

```python
df
```

```python
job_counts.plot(kind = 'bar')
plt.title('Posting by Job Title')
plt.xlabel('')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation=45,ha = 'right') #ha = horizontal alignment
plt.show()
```

```python
df.plot(x = 'job_posted_date', y = 'salary_year_avg')
plt.show()
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()

job_salary.plot(kind = 'barh')
plt.title('Median Salary by Job Title')
plt.xlabel('Salary ($USD)')
plt.ylabel('')
plt.show()
```
